### plt.annotate()
* 값에 주석을 매김
> plt.annotate("주석", xy = (x_value, y_value), xytext(x_location, y_location))

```python
def make_chart_scatter_plot():
    friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
    minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
    labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
    
    plt.scatter(friends, minutes)
    
    #label each point
    for label, friend_count, minute_count in zip(labels, friends, minutes):
        plt.annotate(label, xy = (friend_count, minute_count), xytext=(5,-5), textcoords= 'offset points')
        
    plt.title("Daily Minutes vs. Number of Friends")
    plt.xlabel("# of friends")
    plt.ylabel("daily minutes spent on the site")
    plt.show()
```

```python
make_chart_scatter_plot()
```

### misleading at scatter plot

```python
'''중간고사 점수와 기말고사 점수는 비례하는지 확인하는 chart
misleading 이면 상관관계가 있어보이고, 올바른 차트면 상관관계가 없다.'''
def make_chart_scatterplot_axes(equal_axes=False):
    
    test_1_grades = [99, 90, 85, 97, 80]
    test_2_grades = [100, 85, 60, 90, 70]
    
    plt.scatter(test_1_grades, test_2_grades)
    plt.xlabel("test 1 grade")
    plt.ylabel("test 2 grade")
    
    if equal_axes:
        plt.title("Axes aren't comparable")
        plt.axis("equal")
    else:
        plt.title("misleading: Axes are comparable")
        
    plt.show()
```

```python
#misleading 상관관계가 있어 보인다.
make_chart_scatterplot_axes(False)
```

```python
#상관관계가 없다.
make_chart_scatterplot_axes(True)
```
